```python
# Importação das Bibliotecas
import pandas as pd
import numpy as py
import plotly.express as px
```

```python
# Importando o arquivo Imersão Python - Tabela de ações.xlsx na aba Principal
df_principal = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal # Exibindo todas as linhas
```

```python
# Importando o arquivo Imersão Python - Tabela de ações.xlsx na aba Total_de_acoes
df_total_acoes = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_total_acoes # Exibindo todas as linhas
```

```python
# Importando o arquivo Imersão Python - Tabela de ações.xlsx na aba Ticker
df_ticker = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_ticker # Exibindo todas as linhas
```

```python
# Importando o arquivo Imersão Python - Tabela de ações.xlsx na aba Chatgpt
df_chat_gpt = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Chatgpt")
df_chat_gpt # Exibindo todas as linhas
```

```python
# Realizando uma cópia do dataframe principal apenas com as 4 primeiras colunas
df_principal = df_principal[['Ativo',	'Data',	'Último (R$)','Var. Dia (%)']].copy()
df_principal # Exibindo todas as linhas
```

```python
# Renomeando as colunas 'Último (R$)',	'Var. Dia (%)' e realizando uma cópia do dataframe principal
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal # Exibindo todas as linhas
```

# Cálculos

```python
# Cálculo da Variação percentual do dia 'var_dia_pct'. Nova coluna será a 'Var_pct'
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100 # A variação percentual é o valor inicial divido por cem
df_principal # Exibindo todas as linhas
```

```python
# Cálculo do Valor Inicial em R$. Será calculado a partir do valor final 'valor_final' dividido pela variação percentual 'Var_pct' +1. A nova coluna chamará 'valor_inicial'
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] +1)
df_principal # Exibindo todas as linhas
```

Função `merge() `: permite combinar DataFrames ou objetos da série em um único DataFrame, semelhante à função PROCV do excel.
O merge é usado quando queremos combinar linhas de dois ou mais DataFrames com base em uma chave comum (ou conjunto de chaves), que pode ser um nome de coluna ou um índice. Isso permite juntar informações relacionadas que estão distribuídas em diferentes tabelas, proporcionando uma visão mais completa e útil dos dados.

Merge utilizando método `how = ‘left’` ou `how= ‘right’`

Um merge “left” ou “right” depende de qual tabela você deixa na direita ou esquerda. Para o seguinte cenário faremos um merge do tipo “left”. Mas o mesmo resultado pode ser obtido com um merge “right” trocando a posição das tabelas no método “merge”.

Neste dataset precisando unir os dados da aba 'Principal' com a Planilha 'Total_de_acoes'. Para isso iremos usar a função `merge()`.
df_principal e df_total_acoes. Em cada um dos datasets temos em comum as colunas 'Ativo' e 'Código'.
Usaremos o método how='left', no qual os valores que queremos acrescentar estarão mais a direita e nossos valores principais se manterão à esquerda.

```python
# Função merge() para mesclar as planilhas  df_principal['Ativo'] com a planilha df_total_acoes['Código'], utilizando o método how='left'
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo',right_on='Código',how='left') #left_on='Ativo': a coluna que está na planilha Principal; e right_on='Código': coluna que está na planilha total_acoes
df_principal # Exibindo todas as linhas

# A coluna de interesse a manter é a 'Qtde. Teórica'. A columa código  está duplicada com a coluna 'Ativo'
```

```python
# Removendo a coluna 'Código', que está duplicada
df_principal = df_principal.drop(columns=['Código'])
df_principal # Exibindo todas as linhas
```

```python
# Criando a coluna 'Variacao_rs', que é a Variação em R$ que aquele ativo teve multiplicando a 'Qtde. Teórica'  pela substração do 'valor final' com o 'valor_inicial'
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])* df_principal['Qtde. Teórica']
df_principal # Exibindo todas as linhas
```

```python
# Comando para formatar os dados float
pd.options.display.float_format = '{:.2f}'.format
```

```python
# transformando os dados da coluna 'Qtde. Teórica' em inteiro
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal # Exibindo todas as linhas
```

```python
# Renomeando as colunas 'Qtde. Teórica' e realizando uma cópia do dataframe principal
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal # Exibindo todas as linhas
```

A função `apply()` é uma função que significa recebe uma outra função (neste caso uma função lambda) como parâmetro e a aplica a cada elemento de uma coluna ou linha de um DataFrame.

```python
# Criando a coluna 'Resultado', que é o resultado da Variação em R$ das ações: se elas subiram, desceram ou permaneceram estáveis
# Lamba: cada resultado da coluna Variacao_rs = x
# If 'Variacao_rs' > 0 = Subiu;
# Else 'Variacao_rs' < 0 = Desceu;
# Else 'Estável'
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else'Estável'))
df_principal # Exibindo todas as linhas
```

```python
# Função merge() para mesclar as planilhas  df_principal['Ativo'] com a planilha df_ticker['Ticker'], utilizando o método how='left'
df_principal = df_principal.merge(df_ticker, left_on='Ativo',right_on='Ticker',how='left') #left_on='Ativo': a coluna que está na planilha Principal; e right_on='Ticker': coluna que está na planilha Ticker
df_principal # Exibindo todas as linhas
```

```python
# Removendo a coluna 'Ticker', que está duplicada
df_principal = df_principal.drop(columns=['Ticker'])
df_principal # Exibindo todas as linhas

# A coluna de interesse a manter é a 'Nome'.
```

```python
# Função merge() para mesclar as planilhas  df_principal['Nome'] com a planilha df_chat_gpt['Nome'], utilizando o método how='left'
df_principal = df_principal.merge(df_chat_gpt, left_on='Nome',right_on='Empresa',how='left') #left_on='Nome': a coluna que está na planilha Principal; e right_on='Empresa': coluna que está na planilha chat_gpt
df_principal # Exibindo todas as linhas
```

```python
# Removendo a coluna 'Empresa', que está duplicada
df_principal = df_principal.drop(columns=['Empresa'])
df_principal # Exibindo todas as linhas

# A coluna de interesse a manter é a 'Nome'.
```

```python
# Renomeando as colunas 'Idade (anos)' e realizando uma cópia do dataframe principal
df_principal = df_principal.rename(columns={'Idade (anos)':'Idade'}).copy()
df_principal # Exibindo todas as linhas
```

```python
# Criando a coluna 'Cat_idade', que é o Classificação do tempo que cada empresa existe: 'Menos de 50 anos', 'Entre 50 e 100', 'Mais de 100 anos'
# Lamba: cada resultado da coluna 'Idade' = x
# If 'Idade' > 100 = Mais de 100 anos;
# Else 'Idade' < 50 = Menos de 50 anos;
# Else 'Entre 50 e 100'
df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Menos de 50 anos' if x < 50 else'Entre 50 e 100'))
df_principal # Exibindo todas as linhas
```

# Análises


Foi usado a IA(ChatGPT) para ajudar nos códigos para calcular a partir dos dados calculados da planilha 'Análise' do GoogleSheet:

*Tenho uma planilha com as colunas Variacao_rs e Resultado(Subiu ou Desceu). A partir dessas colunas, monto a atbela abaixo com uso de fórmulas do excel. Se eu tivesse a mesma tabela em um dataframe do pandas como faria para chegar nos mesmos resultados?*

*   Maior variação das Ações do dia
*   Menor variação das Ações do dia

*   Média variação das Ações do dia
*   Média variação das Ações do dia das ações que subiram

*   Média variação das Ações do dia das ações que desceram


```python
# Calculando a maior variação
maior_variacao = df_principal['Variacao_rs'].max()

# Calculando a menor variação
menor_variacao = df_principal['Variacao_rs'].min()

# Calculando a média das variações
media_variacao = df_principal['Variacao_rs'].mean()

# Calculando a média das variações para quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Calculando a média das variações para quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimindo os resultados
print("Maior\tR$", maior_variacao)
print("Menor\tR$", menor_variacao)
print("Média\tR$", media_variacao)
print("Média de quem subiu\tR$", media_subiu)
print("Média de quem desceu\tR$", media_desceu)
```

```python
# Criando um novo dataframe das ações que subiram
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu # Exibindo todas as linhas
```

```python
# Criar um novo dataframe para análise por segmento utilizando a função groupby, onde será realizado a soma (pela função 'sum) por Segmento
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
# reset_index: permanece com o dataframe
df_analise_segmento # Exibindo todas as linhas
```

```python
# Criar um novo dataframe para análise por Resultado utilizando a função groupby, onde será realizado a soma (pela função 'sum) por Variação_rs
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
# reset_index: permanece com o dataframe
df_analise_saldo # Exibindo todas as linhas
```

# Criando Gráficos

```python
# Gráfico de Barras (bar)

fig = px.bar(df_analise_saldo, #selecionando o dataframe desejado
             x='Resultado', # Eixo X
             y='Variacao_rs',  # Eixo Y
             text='Variacao_rs', # Rótulo nas barras, que será os valores de y
             title='Variação Reais por Resultado') # Criando título
fig.show() # Exibindo o gráfico
```

```python
# Gráfico de Barras (bar)

df_analise_saldo['Variacao_rs'] = df_analise_saldo['Variacao_rs'].map("R${:,.0f}".format)

# Definindo o formato para valores em R$
def formatar_reais(x):
    return 'R${:.2f}'.format(x)  # Formato: R$XX,XXX.XX

fig = px.bar(df_analise_saldo, #selecionando o dataframe desejado
             x='Resultado', # Eixo X
             y='Variacao_rs',  # Eixo Y
             text='Variacao_rs', # Rótulo nas barras, que será os valores de y
             title='Variação Reais por Resultado',# Criando título
             color_discrete_sequence=px.colors.qualitative.Set3, #mudando a cor do gráfico
             height=400)  # Altura do gráfico

fig.show() # Exibindo o gráfico






```
